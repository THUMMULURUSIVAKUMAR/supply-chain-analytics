```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load files
df = pd.read_csv(r'C:\Users\HP\Downloads\supply_chain_project\data\tokenized_access_logs.csv', encoding='latin-1')
df_supply = pd.read_csv(r'C:\Users\HP\Downloads\supply_chain_project\data\DataCoSupplyChainDataset.csv', encoding='latin-1')
description = pd.read_csv(r'C:\Users\HP\Downloads\supply_chain_project\data\DescriptionDataCoSupplyChain.csv', encoding='latin-1')

df['Date'] = pd.to_datetime(df['Date'])

print("Access logs shape:", df.shape)
print("Supply chain shape:", df_supply.shape)
print("\nSupply chain columns:", df_supply.columns.tolist())
df_supply.head()
```

```python
print("Shape:", df_supply.shape)
print("\nColumns:\n", df_supply.columns.tolist())
print("\nMissing values:\n", df_supply.isnull().sum())
df_supply.head()
```

```python
top_products = df_supply['Product Name'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values, y=top_products.index, palette='Blues_r')
plt.title('Top 10 Most Demanded Products')
plt.xlabel('Number of Orders')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
dept = df_supply['Department Name'].value_counts()

plt.figure(figsize=(12,5))
sns.barplot(x=dept.index, y=dept.values, palette='Greens_r')
plt.title('Orders by Department')
plt.xlabel('Department')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_supply['order date (DateOrders)'] = pd.to_datetime(df_supply['order date (DateOrders)'])
df_supply['Month'] = df_supply['order date (DateOrders)'].dt.month

monthly_sales = df_supply.groupby('Month')['Sales'].sum()

plt.figure(figsize=(12,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='steelblue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
late = df_supply['Late_delivery_risk'].value_counts()

plt.figure(figsize=(6,5))
plt.pie(late.values, labels=['On Time','Late'], autopct='%1.1f%%',
        colors=['#2ecc71','#e74c3c'], startangle=140)
plt.title('Late Delivery Risk')
plt.tight_layout()
plt.show()
```

```python
cat_profit = df_supply.groupby('Category Name')['Order Profit Per Order'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=cat_profit.values, y=cat_profit.index, palette='Oranges_r')
plt.title('Top 10 Categories by Profit')
plt.xlabel('Total Profit')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
ship = df_supply['Shipping Mode'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=ship.index, y=ship.values, palette='Purples_r')
plt.title('Orders by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Count')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
